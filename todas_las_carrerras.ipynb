```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
carreras = ['CGF','CIV','ECA','IND','MCT','MEC','ELE']
cb = {}
for car in carreras:
    cb[car] = pd.read_csv(f'{car}.csv')
```

```python
for car in carreras:
    sns.set(style="darkgrid")
    sns.histplot(data=cb[car], x="tardoCB",kde=True)
    print(f"{car}:")
    plt.show()
```

```python
carreras = ['CGF','CIV','ECA','MCT','ELE']
for car in carreras:
    corr_df = cb[car].corr()
    plt.figure(figsize=(10,10))
    sns.heatmap(corr_df, annot=True)
    print(f"{car}:")
    plt.show()
```
